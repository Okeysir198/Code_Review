```python
"""
Demo script showing how to use the refactored LLM-driven Call Center Agent.
"""

from call_center_agent import CallCenterAgent
from call_center_enums import CallStep, VerificationStatus, PaymentMethod
from langchain_openai import ChatOpenAI

def create_demo_agent():
    """
    Create a demo agent with sample configuration.
    
    Returns:
        Configured CallCenterAgent instance
    """
    # Create LLM - replace with your preferred LLM
    llm = ChatOpenAI(
        model="gpt-4",  # Can replace with whatever model you prefer
        temperature=0.3
    )
    
    # Sample client details
    client_details = {
        "user_id": 12345,
        "full_name": "John Smith",
        "username": "jsmith",
        "id_number": "8001015001081",
        "email": "john.smith@example.com",
        "vehicle_registration": "ABC123GP",
        "vehicle_make": "Toyota",
        "vehicle_model": "Corolla",
        "vehicle_color": "Silver",
        "outstanding_amount": "R1250.00",
        "subscription_amount": "R250.00",
        "subscription_date": "25th of each month",
        "account_status": "Overdue",
        "cancellation_fee": "R2500.00",
        "total_balance": "R3750.00"
    }
    
    # Configuration for agent
    config = {
        "llm": llm,
        "client_details": client_details,
        "script_type": "ratio_1_inflow",
        "verification": {
            "max_name_verification_attempts": 3,
            "max_details_verification_attempts": 3
        },
        "debug_mode": True,
        "show_logs": True
    }
    
    # Create and return agent
    return CallCenterAgent(config)

def run_interactive_demo():
    """Run an interactive demo with the agent."""
    print("\n=== Call Center Agent Demo ===")
    print("Type 'exit' or 'quit' to end the conversation.")
    print("Agent will simulate a debt collection call for Cartrack.\n")
    
    # Create agent
    agent = create_demo_agent()
    
    # Start conversation
    print("Agent: Hello! This is Cartrack's accounts department. Can I speak with John Smith?")
    
    while True:
        # Get user input
        user_input = input("\nYou: ")
        
        # Check for exit command
        if user_input.lower() in ["exit", "quit"]:
            print("\nDemo ended.")
            break
            
        # Get agent response
        response = agent.run(user_input)
        print(f"\nAgent: {response}")

if __name__ == "__main__":
    run_interactive_demo()
```

```python
import json
import os
import sys
import time # Import the time module
import traceback # For detailed error printing
import pprint # For printing complex objects if JSON fails
from typing import List, Dict, Any, Optional

# Attempt to import your actual application config and agent components
try:
    # Assuming app_config.py is in a location accessible via PYTHONPATH
    from app_config import CONFIG
except ImportError:
    print("Warning: Could not import app_config. Using default mock config.")
    CONFIG = {"client_details": {"full_name": "Test Client"}} # Fallback config

try:
    # Assuming src/Agents/agent00_debtor_call_center.py is accessible
    from src.Agents.agent00_debtor_call_center import (
        CallCenterAgent,
        CallCenterAgentState,
        CallStep,
        VerificationStatus,
        PaymentMethod
    )
    # Import message types from langchain core
    from langchain_core.messages import HumanMessage, AIMessage, SystemMessage, BaseMessage
except ImportError as e:
    print(f"FATAL ERROR: Could not import required Agent components: {e}")
    print("Please ensure the script is run from a location where 'app_config' and 'src' are accessible.")
    sys.exit(1) # Exit if essential components are missing


# Global agent initialization (Using Actual Agent)
# Ensure CONFIG is loaded correctly before this point
try:
    agent_config = CONFIG.copy()
    # Add any necessary default values or overrides for testing if needed
    # e.g., agent_config['llm'] = get_mock_llm_if_needed_for_testing()
    AGENT = CallCenterAgent(agent_config)
    AGENT.show_logs = True  # Enable logging to see more details
    print("Successfully initialized CallCenterAgent.")
except Exception as e:
    print(f"FATAL ERROR: Failed to initialize CallCenterAgent: {e}")
    traceback.print_exc()
    sys.exit(1)


def capture_output(func):
    """
    Decorator to capture stdout and stderr

    Returns:
        Tuple of (original return value, captured stdout, captured stderr)
    """
    def wrapper(*args, **kwargs):
        old_stdout = sys.stdout
        old_stderr = sys.stderr
        import io
        stdout_capture = io.StringIO()
        stderr_capture = io.StringIO()
        try:
            sys.stdout = stdout_capture
            sys.stderr = stderr_capture
            result = func(*args, **kwargs)
            stdout_output = stdout_capture.getvalue()
            stderr_output = stderr_capture.getvalue()
            return result, stdout_output, stderr_output
        finally:
            sys.stdout = old_stdout
            sys.stderr = old_stderr
    return wrapper

def safe_serialize(obj, processed_ids=None):
    """
    Convert non-serializable objects to string representation, handling potential cycles.
    """
    if processed_ids is None:
        processed_ids = set()

    # Prevent infinite recursion on complex objects
    obj_id = id(obj)
    if obj_id in processed_ids:
        return "<Cycle Detected>"
    processed_ids.add(obj_id)

    try:
        if isinstance(obj, (HumanMessage, AIMessage, SystemMessage, BaseMessage)):
             # Use a standard representation for Langchain messages
             return f"[{obj.__class__.__name__} type='{getattr(obj, 'type', 'N/A')}'] {getattr(obj, 'content', '<No Content>')}"
        elif isinstance(obj, CallCenterAgentState):
             # Special handling for the state object if needed, accessing its internal dict
             # Accessing .dict() might be necessary if it's a Pydantic model or similar state manager
             if hasattr(obj, 'dict'):
                 state_dict = obj.dict()
             elif isinstance(obj, dict):
                 state_dict = obj
             else:
                 # Fallback if it's neither dict nor has .dict()
                 state_dict = obj.__dict__ if hasattr(obj, '__dict__') else {}

             # Serialize the state dictionary, excluding messages to avoid large output/recursion
             return {k: safe_serialize(v, processed_ids.copy()) for k, v in state_dict.items() if k != 'messages' and not k.startswith('_')}
        elif hasattr(obj, '__dict__'):
             # Generic handling for other objects
             return {k: safe_serialize(v, processed_ids.copy()) for k, v in obj.__dict__.items() if not k.startswith('_')}
        elif isinstance(obj, (list, tuple)):
            return [safe_serialize(item, processed_ids.copy()) for item in obj]
        elif isinstance(obj, dict):
            return {k: safe_serialize(v, processed_ids.copy()) for k, v in obj.items()}

        # Attempt standard JSON serialization if object is simple enough
        json.dumps(obj)
        result = obj
    except (TypeError, OverflowError):
        # Fallback to string representation for complex/unserializable types
        try:
            result = repr(obj)
        except Exception:
            result = "<Unserializable Object>"
    finally:
        processed_ids.remove(obj_id) # Backtrack after processing

    return result


def pretty_print(title: str, data: Any) -> None:
    """Pretty print test scenario details with safe serialization"""
    print(f"\n{'='*50}")
    print(f"{title}")
    print(f"{'='*50}")
    try:
        safe_data = safe_serialize(data)
        print(json.dumps(safe_data, indent=2, default=str)) # Use default=str for any lingering issues
    except Exception as e:
        print(f"Serialization Error during JSON dump: {e}. Falling back to pprint.")
        pprint.pprint(data) # Fallback to pprint for better complex object display
    print(f"{'='*50}")


def create_test_state(
    messages: List[BaseMessage],
    current_step: Optional[str] = None,
    verification_status: Optional[str] = None,
    additional_context: Optional[Dict[str, Any]] = None
) -> CallCenterAgentState:
    """
    Enhanced helper function to create a test state using the actual CallCenterAgentState.
    """
    # Prepare the initial state dictionary
    initial_state_dict = {"messages": messages}

    if current_step:
        initial_state_dict['current_call_step'] = current_step
    if verification_status:
        initial_state_dict['name_verification_status'] = verification_status # Or appropriate field
    if additional_context:
        initial_state_dict.update(additional_context)

    # Instantiate the actual CallCenterAgentState
    # This assumes CallCenterAgentState can be initialized from a dictionary
    # or has fields that can be set. Adjust if initialization is different.
    try:
        # If CallCenterAgentState inherits from LangGraph's MessagesState or similar Pydantic models:
        state = CallCenterAgentState(**initial_state_dict)
        # OR if it's a simpler class:
        # state = CallCenterAgentState()
        # for key, value in initial_state_dict.items():
        #     if hasattr(state, key):
        #         setattr(state, key, value)
        #     else:
        #         # Handle cases where state object might store things differently (e.g., in a dict attribute)
        #         # This depends heavily on CallCenterAgentState's implementation
        #         pass # Add specific logic if needed based on your state class design
    except Exception as e:
        print(f"Error creating CallCenterAgentState: {e}. Using raw dict as fallback.")
        # Fallback to using a dictionary if state class instantiation fails
        state = initial_state_dict

    return state

@capture_output
def test_scenario(
    scenario_name: str,
    messages: List[BaseMessage],
    current_step: Optional[str] = None,
    verification_status: Optional[str] = None,
    additional_context: Optional[Dict[str, Any]] = None,
    expected: Optional[Dict[str, Any]] = None
) -> Dict[str, Any]:
    """Generic test scenario runner with performance timing"""
    print(f"\n{'='*50}")
    print(f"TEST SCENARIO: {scenario_name}")
    print(f"{'='*50}")

    # Prepare state using the actual State class
    state = create_test_state(
        messages=messages,
        current_step=current_step,
        verification_status=verification_status,
        additional_context=additional_context
    )

    # Print input context
    print("\n--- Input Context ---")
    print("Conversation:")
    for msg in messages:
        # Use the updated safe_serialize which handles actual message objects
        print(safe_serialize(msg))

    print("\n--- State Before Router ---")
    # Use safe_serialize for potentially complex state object
    pretty_print("Initial State", state)


    # --- PERFORMANCE TIMING START ---
    start_time = time.perf_counter()
    # ---

    # Run router
    updates = {} # Initialize updates
    duration_ms = 0.0
    try:
        # Ensure the state object passed is compatible with what AGENT._router expects
        # If _router expects a dictionary, you might need to convert `state` back or adjust _router
        # Assuming _router can handle the CallCenterAgentState object directly:
        updates = AGENT._router(state) # Call the agent's router function

        # --- PERFORMANCE TIMING END ---
        end_time = time.perf_counter()
        duration_ms = (end_time - start_time) * 1000
        # ---

        # Print router results
        print("\n--- Router Analysis ---")
        pretty_print("Router Updates", updates)

        # --- PRINT PERFORMANCE ---
        print("\n--- Performance ---")
        print(f"Router Execution Time: {duration_ms:.2f} ms")
        # ---

        # Verify key outputs
        print("\n--- Verification ---")
        # Adapt keys based on your actual router's output structure
        print(f"Recommended Next Node: {updates.get('recommended_next_node', 'N/A')}")
        print(f"Router Reasoning: {updates.get('router_reasoning', 'N/A')}")


        print("\nâœ“ Router processing completed successfully")

        # Optional validation
        if expected:
            print("\n--- Expectation Validation ---")
            scenario_passed = True
            for key, expected_value in expected.items():
                actual_value = updates.get(key)
                print(f"{key}: Expected='{expected_value}', Got='{actual_value}'")

                # Compare values
                if actual_value != expected_value:
                    print(f"âŒ FAILED: Mismatch in {key}")
                    scenario_passed = False
                else:
                    print(f"âœ… PASSED: {key}")

            print(f"\nScenario '{scenario_name}': {'PASSED âœ“' if scenario_passed else 'FAILED âœ—'}")

    except Exception as e:
        # --- PERFORMANCE TIMING END (on error) ---
        end_time = time.perf_counter()
        duration_ms = (end_time - start_time) * 1000
        # ---
        print(f"\n--- Performance (Error) ---")
        print(f"Time until error: {duration_ms:.2f} ms")
        print(f"\nâŒ Error in router processing: {e}")
        traceback.print_exc() # Print full traceback for debugging
        updates = {"error": str(e), "duration_ms": duration_ms} # Include duration in error case

    # Add duration to the returned results for potential aggregation
    updates['duration_ms'] = duration_ms
    return updates

def generate_test_scenarios():
    """Generate comprehensive test scenarios using actual message types"""
    # Use the actual HumanMessage and AIMessage classes imported
    return [
         # --- Verification Scenarios ---
        {
            "name": "Name Verification - Verified",
             "messages": [
                 AIMessage(content="Good day, this is Cartrack. May I speak to Test Client?"),
                 HumanMessage(content="Yes, speaking.")
             ],
             "current_step": CallStep.INTRODUCTION.value,
             "verification_status": VerificationStatus.INSUFFICIENT_INFO.value,
             "expected": {
                 "recommended_next_node": "name_verification_tool" # Expect router to go to name tool first
             }
         },
        {
            "name": "Details Verification - Sufficient Info Provided",
            "messages": [
                AIMessage(content="Thanks, Can you confirm your vehicle registration?"),
                HumanMessage(content="Yes, it's AB12CD GP")
            ],
            "current_step": CallStep.DETAILS_VERIFICATION.value,
            "verification_status": VerificationStatus.VERIFIED.value, # Assume name verified
            "additional_context": {
                "details_verification_status": VerificationStatus.VERIFIED.value
            },
            "expected": {
                # Router should decide based on verified details state
                "recommended_next_node": "reason_for_call" # Or whatever your router logic dictates
            }
        },
        {
            "name": "Details Verification - Insufficient/Incorrect Info",
            "messages": [
                AIMessage(content="Okay, and what is your email address on file?"),
                HumanMessage(content="I think it's john.doe@example.com")
            ],
            "current_step": CallStep.DETAILS_VERIFICATION.value,
            "verification_status": VerificationStatus.VERIFIED.value, # Name verified
             "additional_context": {
                 "details_verification_status": VerificationStatus.INSUFFICIENT_INFO.value
             },
             "expected": {
                 "recommended_next_node": "details_verification_tool" # Expect router to retry details
             }
        },

         # --- Post-Verification Flow ---
        {
            "name": "Transition from Reason For Call",
            "messages": [
                AIMessage(content="...and an immediate payment of R500 is required."),
                HumanMessage(content="Okay, I understand.")
            ],
            "current_step": CallStep.REASON_FOR_CALL.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                "recommended_next_node": "negotiation"
            }
        },
        {
            "name": "Negotiation - Acknowledging Consequences",
            "messages": [
                AIMessage(content="Failure to pay means services like the app will be suspended."),
                HumanMessage(content="Oh, I didn't realize that would happen.")
            ],
            "current_step": CallStep.NEGOTIATION.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 # Expect router to continue negotiation or move based on sentiment/rules
                 "recommended_next_node": "negotiation" # Or "promise_to_pay" depending on router logic
             }
        },
         {
            "name": "Negotiation - Agreeing to Discount Offer (Pre-Legal)",
            "messages": [
                AIMessage(content="We can offer a 50% discount if paid this month. Can you take this offer?"),
                HumanMessage(content="Yes, that sounds good, I can do that.")
            ],
            "current_step": CallStep.NEGOTIATION.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "promise_to_pay" # Router detects payment intent
             }
        },
        {
            "name": "PromiseToPay - Client Chooses DebiCheck",
            "messages": [
                AIMessage(content="How would you like to arrange the payment? We have DebiCheck, Portal link, or immediate debit."),
                HumanMessage(content="Let's do the DebiCheck option.")
            ],
            "current_step": CallStep.PROMISE_TO_PAY.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 # Router detects payment intent, might stay in PTP or move based on rules
                 "recommended_next_node": "promise_to_pay" # Or "debicheck_setup" depending on exact rules
             }
        },
         {
            "name": "PromiseToPay - Client Chooses Payment Portal",
            "messages": [
                AIMessage(content="How would you like to arrange the payment? We have DebiCheck, Portal link, or immediate debit."),
                HumanMessage(content="Send me the link for the payment portal.")
            ],
            "current_step": CallStep.PROMISE_TO_PAY.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 # Router detects payment intent/request
                 "recommended_next_node": "promise_to_pay" # Or "payment_portal" depending on rules
             }
        },
        {
            "name": "After DebiCheck Setup Explanation",
            "messages": [
                AIMessage(content="...so please approve the request from your bank app."),
                HumanMessage(content="Okay, I will look out for it.")
            ],
            "current_step": CallStep.DEBICHECK_SETUP.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "subscription_reminder"
             }
        },
        {
            "name": "After Payment Portal Guidance",
            "messages": [
                AIMessage(content="...you can click the link and follow the steps to pay online."),
                HumanMessage(content="Got it, thanks.")
            ],
            "current_step": CallStep.PAYMENT_PORTAL.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "subscription_reminder"
             }
        },
          {
            "name": "After Subscription Reminder",
            "messages": [
                AIMessage(content="Just confirming your normal subscription will also debit on the usual date."),
                HumanMessage(content="Alright.")
            ],
            "current_step": CallStep.SUBSCRIPTION_REMINDER.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "client_details_update"
             }
        },
        {
            "name": "After Client Details Update",
            "messages": [
                AIMessage(content="Thanks, I've updated your email address."),
                HumanMessage(content="Okay, great.")
            ],
            "current_step": CallStep.CLIENT_DETAILS_UPDATE.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "referrals"
             }
        },
        {
            "name": "After Referrals (Declined)",
            "messages": [
                AIMessage(content="Do you know anyone who might be interested in Cartrack?"),
                HumanMessage(content="No, not really.")
            ],
            "current_step": CallStep.REFERRALS.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "further_assistance"
             }
        },
          {
            "name": "After Further Assistance (No more questions)",
            "messages": [
                AIMessage(content="Is there anything else I can help with today?"),
                HumanMessage(content="No, that's all, thank you.")
            ],
            "current_step": CallStep.FURTHER_ASSISTANCE.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "closing"
             }
        },
          # --- Query during Negotiation ---
        {
            "name": "Query about Discount Terms during Negotiation",
            "messages": [
                AIMessage(content="We are offering a discount..."),
                HumanMessage(content="How long is that discount valid for?")
            ],
            "current_step": CallStep.NEGOTIATION.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "query_handler" # Router detects query
             }
        },
        # --- Request Cancellation ---
        {
            "name": "Request Cancellation during Negotiation",
            "messages": [
                AIMessage(content="Paying now reinstates your services."),
                HumanMessage(content="Actually, I just want to cancel my service.")
            ],
            "current_step": CallStep.NEGOTIATION.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                "recommended_next_node": "cancellation" # Router detects cancellation request
            }
        },
        # --- Edge Case: Empty/Garbage Input ---
        {
            "name": "Empty Client Message after Verification",
            "messages": [
                AIMessage(content="Okay, your details are verified. The reason for the call is..."),
                HumanMessage(content="") # Empty message
            ],
            "current_step": CallStep.REASON_FOR_CALL.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 "recommended_next_node": "negotiation" # Router applies standard progression
             }
        },
        {
            "name": "Unrelated Statement during Negotiation",
            "messages": [
                AIMessage(content="Paying now reinstates your services."),
                HumanMessage(content="The weather is nice today.")
            ],
            "current_step": CallStep.NEGOTIATION.value,
            "verification_status": VerificationStatus.VERIFIED.value,
            "expected": {
                 # Router should apply standard progression or re-engage based on its rules
                 "recommended_next_node": "negotiation" # Example: try to re-engage
             }
        },

    ]

def run_comprehensive_tests():
    """Run comprehensive router tests and collect performance data"""
    print("\n=== Running Comprehensive Router Tests ===")

    scenarios = generate_test_scenarios()
    results_summary = []
    total_duration = 0
    test_count = 0

    for i, scenario in enumerate(scenarios):
        print(f"\n--- Starting Scenario {i+1}: {scenario['name']} ---")
        # Prepare scenario details
        test_params = {
            "scenario_name": scenario["name"],
            "messages": scenario.get("messages", []),
            "current_step": scenario.get("current_step"),
            "verification_status": scenario.get("verification_status"),
            "additional_context": scenario.get("additional_context"),
            "expected": scenario.get("expected")
        }

        # Run test scenario with output capture
        result, stdout, stderr = test_scenario(**test_params)

        # Store result summary
        duration = result.get('duration_ms', 0)
        scenario_passed = result.get('error') is None and (
            not scenario.get("expected") or
            result.get("recommended_next_node") == scenario["expected"].get("recommended_next_node")
        )

        results_summary.append({
            "scenario": scenario["name"],
            "recommended_node": result.get("recommended_next_node", "ERROR"),
            "reasoning": result.get("router_reasoning", "N/A"),
            "duration_ms": duration,
            "passed_validation": scenario_passed,
            "error": result.get("error")
        })

        if result.get('error') is None:
            total_duration += duration
            test_count += 1
        else:
             print(f"ERROR recorded for scenario: {result.get('error')}")


        # Print captured stdout/stderr if needed for debugging LLM calls within router
        # if stdout:
        #      print(f"\n--- Captured STDOUT for Scenario {i+1} ---")
        #      print(stdout)
        # if stderr:
        #      print(f"\n--- Captured STDERR for Scenario {i+1} ---")
        #      print(stderr)
        print(f"\n{'='*20} End Scenario {i+1} {'='*20}\n")


    print("\n=== Router Test Summary ===")
    passed_count = sum(1 for r in results_summary if r['passed_validation'])
    failed_count = len(results_summary) - passed_count

    print(f"Total Scenarios: {len(results_summary)}")
    print(f"Passed: {passed_count}")
    print(f"Failed: {failed_count}")

    if test_count > 0:
        average_duration = total_duration / test_count
        print(f"\nAverage Router Execution Time (Successful Runs): {average_duration:.2f} ms")
    else:
        print("\nNo successful runs to calculate average time.")

    # Optionally print detailed results
    print("\n--- Detailed Results ---")
    for res in results_summary:
         status = "âœ… PASSED" if res['passed_validation'] else "âŒ FAILED"
         print(f"[{status}] Scenario: {res['scenario']}")
         print(f"  -> Recommended Node: {res['recommended_node']}")
         print(f"  -> Reasoning: {res['reasoning']}")
         print(f"  -> Duration: {res['duration_ms']:.2f} ms")
         if res['error']:
             print(f"  -> Error: {res['error']}")


    print("\n=== All Comprehensive Router Tests Completed ===")

if __name__ == "__main__":
    # Basic check to ensure AGENT is initialized before running tests
    if 'AGENT' in locals() and AGENT is not None:
        run_comprehensive_tests()
    else:
        print("Agent initialization failed. Cannot run tests.")
```
